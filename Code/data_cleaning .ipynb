```python
import pandas as pd
```

```python
df_match = pd.read_csv(r"C:\Users\sumit\OneDrive\Desktop\WorldCup2022\raw_data\match_result.csv")
```

**match_result**

```python
df_match
df_match = df_match.replace("U.A.E.","United Arab Emirates")
```

```python
match_id = {}

for index,row in df_match.iterrows():
    key1 = row["Team 1"] + ' VS ' + row["Team 2"]
    key2 = row["Team 2"] + ' VS ' + row["Team 1"]
    
    match_id[key1]  = row["id"]
    match_id[key2]  = row["id"]
    
match_id
```

**batting summery**

```python
batting_summery = pd.read_csv(r"C:\Users\sumit\OneDrive\Desktop\WorldCup2022\raw_data\battingSummery.csv")
batting_summery[5:20]
```

```python
# creating out and not out column

batting_summery["out/not_out"] = batting_summery["dismissal"].apply(lambda x : "not out" if x=="not out" else "out" )
```

```python
# removing the unnecessary words from the data

batting_summery["batsman"] = batting_summery["batsman"].apply(lambda x: x.replace("†",""))
# batting_summery["batsman"] = batting_summery["batsman"].apply(lambda x: x.replace("(c)"),"")
# batting_summery["batsman"] = batting_summery["batsman"].apply(lambda x: x.replace("(wk)"),"")
```

```python
# data Modelling for batting

batting_summery["id"] = batting_summery["team"].map(match_id)

batting_summery
```

```python
batting_summery.to_csv(r"C:\Users\sumit\OneDrive\Desktop\WorldCup2022\cleaned_data\batting_summery.csv",index=False)
```
